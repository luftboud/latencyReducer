```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("webrtc_stats_1776190751821.csv")
df = df[:20]
```

```python
df["time"] = pd.to_datetime(df["time"])
df["elapsed_s"] = (df["time"] - df["time"].iloc[0]).dt.total_seconds()
```

```python
colors = ["#580F41", "#800080", "#00008B", "#030764", "#01153E", "#0000FF", "#0343DF"] 
```

```python
def plot(metric, name, color):
    plt.figure(figsize=(10, 5))
    plt.plot(df["elapsed_s"], df[metric], marker="o", color=color)
    plt.xlabel("Time (seconds)")
    plt.ylabel(metric)
    plt.title(f"{name} over time")
    plt.xticks(df["elapsed_s"])
    plt.tight_layout()
    plt.savefig(f"plots/{metric}.png")

    plt.close()

metrics = {"rtt": "Round-trip time", 
           "bitrate": "Bitrate (Kbps)", 
           "fps": "Frames per second", 
           "receivedPackets": "Packets received by client",
           "jitter": "Packet delay variation",
           "buffer": "Average time(ms) spent in jitter buffer by 1 frame", 
           "decode": "Frames decoded correctly"}

for (metric, descr), color in zip(metrics.items(), colors):
    plot(metric, descr, color)
```
